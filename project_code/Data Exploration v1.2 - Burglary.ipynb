```julia
# clean data remove all NAs
crime = crime[!isna(crime[:Lon]) & !isna(crime[:Lat]) & !isna(crime[:Police_Districts]) & !isna(crime[:UCR_General]), :];
crime = sort(crime, cols = :Dispatch_Date)
```

```julia
crime[:Dispatch_Date_Time] = DateTime(crime[:Dispatch_Date_Time],"yyyy-mm-dd HH:MM:SS")
```

```julia
crime[:Month] = DateTime(crime[:Month],"yyyy-mm")
```

```julia
crime_pivot = by(crime,[:Month,:Text_General_Code],nrow)
```

```julia
types_of_crime = by(crime,[:Text_General_Code],nrow)
types_of_crime = sort(types_of_crime,cols = :x1,rev=true)
```

```julia
# there are too many small category of Thefts, we want to combine them together
Thefts1 = crime_pivot[crime_pivot[:Text_General_Code].=="Thefts",:]
Thefts2 = crime_pivot[crime_pivot[:Text_General_Code].=="Theft from Vehicle",:]
Thefts3 = crime_pivot[crime_pivot[:Text_General_Code].== "Motor Vehicle Theft",:]
Theft = vcat(Thefts1,Thefts2,Thefts3)
Theft = by(Theft,[:Month],df -> sum(df[:x1]))
Theft[:Text_General_Code] =  fill("Thefts",nrow(Theft),1)[:,1]
Theft
```

```julia
# And combine "Burglary Residential",  "Burglary Non-Residential" to "Burglary"
Burglary1 = crime_pivot[crime_pivot[:Text_General_Code].=="Burglary Residential",:]
Burglary2 = crime_pivot[crime_pivot[:Text_General_Code].=="Burglary Non-Residential",:]
Burglary = vcat(Burglary1,Burglary2)
Burglary = by(Burglary,[:Month],df -> sum(df[:x1]))
Burglary[:Text_General_Code] =  fill("Burglary",nrow(Burglary),1)[:,1]
Burglary
```

```julia
# Combine "Robbery Firearm",    "Robbery No Firearm" to "Robbery"
Robbery1 = crime_pivot[crime_pivot[:Text_General_Code].=="Robbery Firearm",:]
Robbery2 = crime_pivot[crime_pivot[:Text_General_Code].=="Robbery No Firearm",:]
Robbery = vcat(Robbery1,Robbery2)
Robbery = by(Robbery,[:Month],df -> sum(df[:x1]))
Robbery[:Text_General_Code] =  fill("Robbery",nrow(Robbery),1)[:,1]
Robbery
```

```julia
# Combine "Public Drunkenness", "Liquor Law Violations" to "Drinking Violation"
Drinking1 = crime_pivot[crime_pivot[:Text_General_Code].=="Public Drunkenness",:]
Drinking2 = crime_pivot[crime_pivot[:Text_General_Code].=="Liquor Law Violations",:]
Drinking = vcat(Drinking1,Drinking2)
Drinking = by(Drinking,[:Month],df -> sum(df[:x1]))
Drinking[:Text_General_Code] =  fill("Drinking Violation",nrow(Drinking),1)[:,1]
Drinking
```

```julia
# Combine "Homicide - Justifiable", "Homicide - Criminal" , "Homicide - Gross Negligence" into "Homicide"
Homicide1 = crime_pivot[crime_pivot[:Text_General_Code].=="Homicide - Justifiable",:]
Homicide2 = crime_pivot[crime_pivot[:Text_General_Code].=="Homicide - Criminal" ,:]
Homicide3 = crime_pivot[crime_pivot[:Text_General_Code].=="Homicide - Gross Negligence",:]
Homicide = vcat(Homicide1,Homicide2,Homicide3)
Homicide = by(Homicide,[:Month],df -> sum(df[:x1]))
Homicide[:Text_General_Code] =  fill("Homicide",nrow(Homicide),1)[:,1]
Homicide
```

```julia
# Combine "Aggravated Assault No Firearm", "Aggravated Assault Firearm" into "Aggravated Assault"
Aggravated1 = crime_pivot[crime_pivot[:Text_General_Code].=="Aggravated Assault No Firearm",:]
Aggravated2 = crime_pivot[crime_pivot[:Text_General_Code].=="Aggravated Assault Firearm" ,:]
Aggravated = vcat(Aggravated1,Aggravated2)
Aggravated = by(Aggravated,[:Month],df -> sum(df[:x1]))
Aggravated[:Text_General_Code] =  fill("Aggravated Assault",nrow(Aggravated),1)[:,1]
Aggravated
```

```julia
# combine the whole categorized dataset
othercrimes = (crime_pivot[:Text_General_Code] .!= "Thefts") &
            (crime_pivot[:Text_General_Code] .!= "Theft from Vehicle") & 
            (crime_pivot[:Text_General_Code] .!=  "Motor Vehicle Theft" ) &
            (crime_pivot[:Text_General_Code] .!= "Burglary Residential") &
            (crime_pivot[:Text_General_Code] .!= "Burglary Non-Residential"    ) &
            (crime_pivot[:Text_General_Code] .!= "Robbery Firearm") &
            (crime_pivot[:Text_General_Code] .!=  "Robbery No Firearm" ) &
            (crime_pivot[:Text_General_Code] .!= "Public Drunkenness") &
            (crime_pivot[:Text_General_Code] .!=  "Liquor Law Violations" ) &
            (crime_pivot[:Text_General_Code] .!=  "Homicide - Justifiable" ) &
            (crime_pivot[:Text_General_Code] .!=  "Homicide - Criminal") &
            (crime_pivot[:Text_General_Code] .!= "Homicide - Gross Negligence" ) &
            (crime_pivot[:Text_General_Code] .!=  "Aggravated Assault Firearm") &
            (crime_pivot[:Text_General_Code] .!= "Aggravated Assault No Firearm" ) 
othercrimes = crime_pivot[othercrimes.==true,:]
crime_cleaned = vcat(othercrimes,Theft,Burglary,Robbery,Drinking,Homicide,Aggravated)
crime_cleaned = sort(crime_cleaned,cols =:Month,rev=true)
```

```julia
types_of_crime_cleaned = by(crime_cleaned,[:Text_General_Code],df -> sum(df[:x1]))
types_of_crime_cleaned = sort(types_of_crime_cleaned,cols = :x1,rev=true)
```

#  Fitting Part
First we want to just look at Thefts, the most common type of crime. We will also start with a rather simply model. It seems from the graph that there exist some seasonal trend fluctrating similarly each year. The first thought we had was let X be month number. However, the correlation between month number (increasing from Jan to Dec) and number of incidents (seasonally flucturating up and down) does not appear to be linear or polynomial. However, if we plot it against temperature, there seems to be a linear correlation.

```julia
n = nrow(DBurglary)
X₁ = [ones(n) DBurglary[:Temperature_Avg].data]
y₁ = DBurglary[:x1].data
w₁ = X₁\y₁
```

```julia
DBurglary[:AvgTempModel] = X₁*w₁
plot(DBurglary,x=:Temperature_Avg,y=:x1,Geom.point,layer(x=:Temperature_Avg,y=:AvgTempModel,Geom.line))
```

```julia
plot(DBurglary,x=:Month,y=:x1,Geom.point,layer(x=:Month,y=:AvgTempModel,Geom.line),
Guide.title("Fitting of Burglary vs. Temp"))
```

The temperature seems like a good indicator for the number of incidents in this month. However, it seems like the model predicted 2009 criminal rate far above actual.  With this information, we also want to examine if we could add another predicator, Number of incidents 12 months before. From the plot below, there seems to be some positive linear correlation.

```julia
year_later = DBurglary[:x1][13:end]
year_before = DBurglary[:x1][1:(end-12)]
plot(x=year_before,y=year_later,Geom.point,
Guide.ylabel("Number Of Incidents"), 
Guide.xlabel("Number Of Incidents 12 Month Before"),
Guide.title("Burglary Incidents vs. 12 Month Before"))
```

```julia
n₂ = size(year_later)
X₂ = [ones(n₂) DBurglary[:Temperature_Avg].data[13:end] year_before.data]
y₂ = year_later.data
w₂ = X₂\y₂
```

```julia
set_default_plot_size(20cm,9cm)
DBurglary[:Temp12MonthModel] = vcat(DBurglary[:AvgTempModel][1:12],X₂*w₂)
plot(DBurglary,x=:Month,y=:x1,
Geom.point,layer(x=:Month,y=:Temp12MonthModel,Geom.line,Theme(default_color=colorant"orange")),
layer(x=:Month,y=:AvgTempModel,Geom.line),
Guide.manual_color_key("Legend", ["Model With Temp", "Model With Temp and 12 Month Before"], ["deepskyblue", "orange"]))
```

From this above graph, it seems like the updated model with Temp and 12 month before's data actually did even worse for 2010, due to the low rate for 2009.  

```julia
err_temp = sum((y₁-X₁*w₁).^2)
err_temp_olddata = sum((y₂-X₂*w₂).^2)
[err_temp err_temp_olddata]
```

However, if we look at error on training set, the updated model has a small in sample error (which make sense because we are fitting in more information)

Now, we also want to see if any of the indicators have a big impact on our model. From a brief look at the annual indicators, it seems like unemployment rate for 2009 is especially low, and number of day of good air is especially high for 2009. We wonder if they could explain the drop in crime rate that year. 

```julia
X₃= [ones(n) DBurglary[:Temperature_Avg].data DBurglary[:Unemployment].data DBurglary[:Days_good_air_quality].data]
y₃ = DBurglary[:x1].data
w₃ = X₃\y₃
```

```julia
DBurglary[:TempIndicators] = X₃*w₃ 
plot(DBurglary,x=:Month,y=:x1,
Geom.point,layer(x=:Month,y=:Temp12MonthModel,Geom.line,Theme(default_color=colorant"orange")),
layer(x=:Month,y=:AvgTempModel,Geom.line),
layer(x=:Month,y=:TempIndicators,Geom.line,Theme(default_color=colorant"purple")),
Guide.manual_color_key("Legend", ["Model With Temp", "Model With Temp and 12 Month Before","Model With Temp and Indicators"], ["deepskyblue", "orange","purple"]))
```

Interestingly, this new model does not seem to do a lot better than the first model just from looking at the graph. Let's look at the error.

```julia
err_temp_indicators = sum((y₃-X₃*w₃).^2)
[err_temp err_temp_olddata err_temp_indicators]
```

The square error is actually higher for than the second model, and not remarkably smaller than the first model
